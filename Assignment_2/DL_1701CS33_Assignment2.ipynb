```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('https://github.com/piyushchauhan/DL-CS551/raw/master/Assignment_2/diabetes.csv')
```

```python
df.head()
```

```python
print(df.describe().T)
X = df.iloc[:,:-1] 
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, np.array(tf.one_hot(y, 2)), test_size=0.2, random_state=44, shuffle =True)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(8,)),
    tf.keras.layers.Dense(500, activation='sigmoid'),
    tf.keras.layers.Dense(100, activation='sigmoid'),
    tf.keras.layers.Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
BATCH_SIZE = 70
EPOCHS = 1000
```

```python
history = model.fit(x=X_train,y=y_train,batch_size=BATCH_SIZE,epochs=EPOCHS,)
```

```python
y_pred = model.predict(X_test)
```

```python
a = np.argmax(y_pred, axis=1)
print(a)
print(a.shape)
print(a[0])
```

```python
b = np.argmax(y_test, axis=1)
print(b)
print(b.shape)
print(b[0])
```

```python
print(classification_report(a,b))
```
